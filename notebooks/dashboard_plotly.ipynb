```python
# setup inicial
import pandas as pd

df = pd.read_csv("../data/ecom_data.csv")

df["Data_Venda"] = pd.to_datetime(df["Data_Venda"])
df["Valor_Total"] = df["Valor_Unitario"] * df["Quantidade"]

df.head()
```

```python
faturamento = df["Valor_Total"].sum()
ticket = df["Valor_Total"].mean()
pedidos = df["ID_Transacao"].nunique()

print(faturamento, ticket, pedidos)
```

```python
df["Mes"] = df["Data_Venda"].dt.to_period("M").astype(str)

vendas_mes = df.groupby("Mes")["Valor_Total"].sum().reset_index()
```

```python
import plotly.express as px

fig = px.line(vendas_mes, x="Mes", y="Valor_Total", title="Vendas ao longo do tempo")
fig.show()
```
